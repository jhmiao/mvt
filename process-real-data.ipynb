```python
# turn all zip codes into strings
zip_codes_1 = [str(zip_code) for zip_code in zip_codes_1]
zip_codes_2 = [str(zip_code) for zip_code in zip_codes_2]
# zip_codes_2
```

```python
# drop duplicate values in zip_codes_1 and zip_codes_2
zip_codes_3 = list(set(zip_codes_1))
zip_codes_4 = list(set(zip_codes_2))
```

```python
len(zip_codes_3), len(zip_codes_4)
```

```python
# Function to process a batch
def batch_distance_matrix(origins, destinations, mode="driving"):
    result_matrix = []
    for origin in origins:
        try:
            result = gmaps.distance_matrix(origins=[origin], destinations=destinations, mode=mode)
            distances = [
                element["duration"]["value"] / 60 if element["status"] == "OK" else None
                for element in result["rows"][0]["elements"]
            ]
            result_matrix.append(distances)
        except Exception as e:
            print(f"Error with origin {origin}: {e}")
            result_matrix.append([None] * len(destinations))
    return result_matrix

# Helper function to chunk lists
def chunk_list(input_list, chunk_size):
    for i in range(0, len(input_list), chunk_size):
        yield input_list[i:i + chunk_size]

# Break `zip_codes_1` and `zip_codes_2` into chunks of 25 or fewer
zip_chunks_1 = list(chunk_list(zip_codes_3, 25))
zip_chunks_2 = list(chunk_list(zip_codes_4, 25))

# Initialize the final DataFrame for storing results
travel_time_df = pd.DataFrame(index=zip_codes_3, columns=zip_codes_4)

# Process each chunk
for origins in zip_chunks_1:
    for destinations in zip_chunks_2:
        # Get the travel time matrix for the current chunk
        batch_matrix = batch_distance_matrix(origins, destinations, mode="driving")
        # Convert to DataFrame for easy integration
        batch_df = pd.DataFrame(batch_matrix, index=origins, columns=destinations)
        # Update the main DataFrame
        travel_time_df.update(batch_df)

# Save the travel time matrix to a CSV file (optional)
travel_time_df.to_csv("travel_time_matrix.csv")

# Display the first few rows of the DataFrame
print(travel_time_df.head())
```

```python
# Create an empty 119x62 DataFrame for the full matrix
full_matrix = pd.DataFrame(index=zip_codes_1, columns=zip_codes_2)

# Fill in the matrix with travel times from the unique travel_time_df
for i, origin in enumerate(zip_codes_1):
    for j, destination in enumerate(zip_codes_2):
        full_matrix.iloc[i, j] = travel_time_df.loc[origin, destination]

# Reset the index and columns for duplicates to maintain their original structure
full_matrix.index = zip_codes_1
full_matrix.columns = zip_codes_2

full_matrix.head()
```
